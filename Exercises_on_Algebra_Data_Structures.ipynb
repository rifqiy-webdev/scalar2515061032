```python
import numpy as np
import torch
import tensorflow as tf
```

```python
#selesaikan soal berikut
#what is the transpose of this vector ([[25],[2],[-3],[-23]])
```

```python
#numpy
x = np.array([[25],[2],[-3],[-23]])
x = x.T
x
```

```python
#pytoch
x_pt = torch.tensor([[25],[2],[-3],[-23]])
x_pt = x_pt.T
x_pt
```

```python
#tensorflow
x_tf = tf.Variable([[25],[2],[-3],[-23]])
x_tf = tf.transpose(x_tf)
x_tf
```
