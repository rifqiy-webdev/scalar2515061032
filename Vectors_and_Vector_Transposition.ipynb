```python
# 3. lakukan soal nomor 2 menggunakan pytorch dan tensorflow
import torch
import tensorflow as tf
```

```python
#lakukan transpose pada vector x dan simpan pada variabel x_t
#pytoch
x_pt = torch.tensor([[32, 3, 2]])

x_t=x_pt.T
x_t

```

```python
#tampilkan ukuran dari vector x_t
x_t.shape
```

```python
#lakukan transpose pada vector y dan simpan pada variabel y_t
y_pt = torch.tensor([[23, 2, 3]])

y_t=y_pt.T
y_t
```

```python
#tampilkan ukuran dari vector x_y
y_t.shape
```

```python
#lakukan transpose pada vector x dan simpan pada variabel x_t
#tensorflow

x_tf = tf.Variable([[32, 3, 2]])

x_t = tf.transpose(x_tf)
x_t
```

```python
#tampilkan ukuran dari vector x_t
x_t.shape
```

```python
#lakukan transpose pada vector y dan simpan pada variabel y_t
y_tf = tf.Variable([[23, 2, 3]])

y_t = tf.transpose(y_tf)
y_t
```

```python
#tampilkan ukuran dari vector x_y
y_t.shape
```
